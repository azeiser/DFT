# Diskrete Fourier Transformation

Import von Bibliotheken

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Beispielfunktion
Gegeben sei die 2-periodische Funktion $$f(t) = t\cdot (2-t), \quad 0\le t \le 2.$$

```python
def f(t):
    t_mod = np.mod(t,2)
    return t_mod * (2-t_mod)
```

Diese Funktion wollen wir graphisch darstellen. Dazu erzeugen wir uns einen Vektor mit 1001 Werten zwischen -3 und 5.

```python
tfull = np.linspace(-3,5,1001)
```

An diesen Stellen werten wir die Funktion aus und speichern die Werte in einem Vektor.

```python
yfull = f(tfull)
```

Jetzt k√∂nnen wir den Graph plotten.

```python
plt.figure()
plt.plot(tfull,yfull)
plt.show()
```

## Sampling
Jetzt wollen wir die Funktion im Intervall $[0,2]$ abtasten. Dazu teilen wir das Intervall in 8 gleich gro√üe Teilst√ºcke und nehmen die 9 Endpunkte:

```python
t = np.linspace(0,2,9) # 9 Punkte zwischen 0 und 2
print(t)
```

Wir bestimmen die Funktionswerte an diesen Punkten

```python
y = f(t)
print(y)
```

Diese Punkte k√∂nnen wir zusammen mit der ursrp√ºnglichen Funktion darstellen:

```python
plt.figure()
plt.plot(tfull,yfull, label="Funktion")
plt.scatter(t,y, c="red", label="Abtastung")
plt.xlim((-1,3))
plt.legend()
plt.show()
```

Wir sehen, dass der neunte Datenpunkt wegen der Periodizit√§t gleich dem ersten ist. Daher reicht es nur die ersten 8 Werten zu nehmen. Wir definieren daher

```python
dt = 2.0/8.0
t = np.arange(0,8) * dt # Vektor [0, 1, 2, ..., 7] mit dt multipliziert
print(t)
```

und werten die Funktion an diesen Punkten aus:

```python
y = f(t)
print(y)
```

Graphisch auf einer Periode dargestellt:

```python
plt.figure()
plt.plot(tfull,yfull, label="Funktion")
plt.scatter(t,y, c="red", label="Abtastung")
plt.xlim((0,2))
plt.legend()
plt.show()
```

Insgesamt gen√ºgt es nach dem Sampling, den Vektor y zu kennen, d.h. den Wert der Funktion an 8 Punkten.
